# Model

```python
country_dict = {
    'Kenya': 1,
    'Rwanda': 2,
    'Tanzania': 3,
    'Uganda': 4,
}
df['country'] = df['country'].map(country_dict)
```

```python
bank_dict = {
    'Yes': 1,
    'No': 0
}
df['bank_account'] = df['bank_account'].map(bank_dict)
df['cellphone_access'] = df['cellphone_access'].map(bank_dict)
```

```python
gender_dict = {
    'Male': 1,
    'Female': 0
}
df['gender_of_respondent'] = df['gender_of_respondent'].map(gender_dict)
```

```python
relan_dic = {
    'Head of Household': 1,
    'Spouse': 2,
    'Child': 3,
    'Parent':4,
    'Other relative' : 5,
    'Other non-relatives' : 6
}
df['relationship_with_head'] = df['relationship_with_head'].map(relan_dic)
```

```python
marital_dict = {
    'Married/Living together': 1,
    'Divorced/Seperated': 2,
    'Widowed': 3,
    'Single/Never Married': 4,
    'Dont know' : 5
}
df['marital_status'] = df['marital_status'].map(marital_dict)
```

```python
educa_dict = {
    'No formal education':1,
    'Primary education':2,
    'Secondary education':3,
    'Vocational/Specialised training':4,
    'Tertiary education':5,
    'Other/Dont know/RTA':6
}
df['education_level'] = df['education_level'].map(educa_dict)
```

```python
df['job_type'].unique()
```

```python
job_dic = {
    'Self employed': 1,
    'Government Dependent': 2,
       'Formally employed Private': 3,
    'Informally employed':4,
       'Formally employed Government' : 5,
     'Farming and Fishing' : 6,
       'Remittance Dependent' : 7,
     'Other Income' : 8,
       'Dont Know/Refuse to answer' : 9,
     'No Income': 10
}
df['job_type'] = df['job_type'].map(job_dic)
```

```python
loc_dict = {
    'Urban': 1,
    'Rural': 0
}
df['location_type'] = df['location_type'].map(loc_dict)
```

```python
x = df.drop(['bank_account', 'uniqueid'], axis=1)
y = df['bank_account']
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
